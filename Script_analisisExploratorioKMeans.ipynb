```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


```

# Creacion del Diccionario de datos (DD)

```python
dd=pd.DataFrame()
dd['Nombre']=None
dd['Desc']=None
uno=['country','child_mort','exports','health','imports','income','inflation','live_expect','total_fer','gdpp']
dos=['numero de paises',
     'mortalidad de niños menores de 5 años de edad por cada 1000 nacidos',
     'exportaciones de bienes y servicios per capita. dado por la edad como tambien del pib per cápita',
     'gasto total en salud per capita. dado por la edad como tambien del pib per cápita',
     'importaciones de bienes y servicios per capita. dado por la edad como tambien del pib per cápita',
     'ingresos netos por persona',
     'medida anual tasa de crecimiento del pib total',
     'el promedio en numero de años de vida que un niño recién nacido vivirá, en caso que los patrones de mortalidad sigan siendo los mismos',
     'el numero de niños que nacería, si las tasas de fertilidad actual por mujer se mentienen',
     'el pib per capita, calculado tomando en cuenta el pib total por el total de la poblacion']
dd['Nombre']=uno
dd['Desc']=dos
dd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Jorge2018/RepositorioArchivos/main/Country-data%20-%20Country-data.csv', index_col='country')

df.head()
```

# Revision de datos y comprobacion de duplicados

```python
print('Number of Duplicated Rows', df.duplicated().sum())
print('\n')
print(df.info())
print('\n')
df.describe()

```

# Escalamiento y modelado de los datos

```python
# Scale the data
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
X_scaled.head()

```

```python
# Iterate over a range of k clusters to find the optimal number of clusters
ks = range(2,11)
sils = []
inertias = []
for k in ks:
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(X_scaled)
  sils.append(silhouette_score(X_scaled, kmeans.labels_))
  inertias.append(kmeans.inertia_)
#plot inertias and silhouette scores for each number of clusters.
fig, axes = plt.subplots(1,2, figsize=(15,5))
axes[0].plot(ks, sils)
axes[0].set_title('Silhouette Scores')
axes[0].set_xticks(ks)
axes[1].plot(ks, inertias)
axes[1].set_title('Inertia')
axes[1].set_xticks(ks);

```

*Podemos observar que agrupar los datos en 4 grupos nos da la mejor puntuación de silueta y podemos detectar un codo en el trazado de inercia en 4.  Si bien el trazado de inercia no tiene un codo muy pronunciado en ningún número, la puntuación de silueta parece favorecer los grupos 4 o 5.  Escogeremos el 4. Ajustemos un nuevo modelo con 4 clústeres y agreguemos esos clústeres a una nueva columna en nuestro dataframe original (sin escala)*

```python
# Refit KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_scaled)
# Add the clusters as a column in the dataframe
df['cluster'] = kmeans.labels_
df.head()

```

#Analizar los clústeres

*Ahora que hemos agrupado los datos, podemos analizar los clústeres.  Queremos obtener algunas estadísticas resumidas sobre cada clúster.  ¿Qué estadística queremos?*  

*Los valores medios de cada característica serán los más útiles para entender los clústeres en su conjunto.  Recuperar los valores máximos o mínimos puede ser interesante en algunos casos, pero los valores medios nos dan la idea del clúster en su conjunto.*

*Agruparemos los datos por clúster y agregaremos los grupos por los valores medios de cada característica.*

```python
# Group the dataframe by cluster and aggregate by mean values
# Add as_index=False so that the clusters are a column, not the index
cluster_groups = df.groupby('cluster', as_index=False).mean()
cluster_groups
```

#Visualizar los clústeres

*Como científicos de datos, lidiamos mucho con los números, pero trazar valores es ideal para la comparación e importante para las presentaciones.  Ahora que tenemos nuestros datos agrupados y los valores medios para cada característica calculados, Pandas y Matplotlib hacen que sea bastante sencillo trazar nuestros datos.*

```python
# Visualize means 
fig, axes = plt.subplots(3,4, figsize = (20, 10))
# fig is an array of dimensions 3,4.  axes.ravel() will flatten it to 12,1
axes = axes.ravel()
# Loop over columns and plot each in a separate figure, skip 'cluster' column
for i, col in enumerate(cluster_groups.columns[1:]):
  axes[i].bar(cluster_groups['cluster'], cluster_groups[col])
  axes[i].set_title(f'Mean {col}')

```
