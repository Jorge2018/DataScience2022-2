```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
school_df = pd.read_csv(r'https://raw.githubusercontent.com/Jorge2018/RepositorioArchivos/main/School_Details.csv')

school_df.head(2)
```

```python
school_df.info()
```

*Creacion de bbdd y conection*

```python
# create sqlite engine for a database called schools
engine = create_engine('sqlite:///schools.db', echo=True)
# create connection to engine
conn = engine.connect()
```

*Agregacion de tb a la bbdd recien creada*

```python
# add the dataframe as a table in sqlite
sqlite_table = "school_details"
school_df.to_sql(sqlite_table, conn, if_exists='fail')
```

# R1

```python
# Write the SQL query
sql = """SELECT EstablishmentName, SchoolLevel, Website
 FROM school_details; 
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)

```

# R2

```python
# Write the SQL query
sql = """SELECT EstablishmentName, `DistrictHQDistance (in Km)`
 FROM school_details WHERE `DistrictHQDistance (in Km)` > 100 ; 
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

# R3

```python
sql = """SELECT EstablishmentName, SchoolLevel, `DistrictHQDistance (in Km)`
 FROM school_details WHERE SchoolLevel='JHS' ; 
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

# R4

```python
sql = """SELECT EstablishmentName, SchoolLevel, `DistrictHQDistance (in Km)`
 FROM school_details WHERE EstablishmentName LIKE '%JUNIOR%' OR EstablishmentName LIKE '%HIGH%'  ; 
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

# R5

```python
sql = """SELECT EstablishmentName, SchoolLevel, `DistrictHQDistance (in Km)`
 FROM school_details WHERE EstablishmentName LIKE 'C%' 
 ORDER BY EstablishmentName ASC
 LIMIT 10 ; 
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

# R6

```python
sql = """SELECT EstablishmentName
 FROM school_details WHERE LocatedInRuralOrUrban ='Urban Area'
 ORDER BY EstablishmentName DESC
 ; 
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

# PROMEDIO

```python
sql = """SELECT AVG(`CRCDistance (in Km)`) as PROM_CRC_DISTANCE
FROM school_details;
"""
pd.read_sql_query(sql, engine)

```

# MAX

```python
sql = """SELECT MAX(`DistrictHQDistance (in Km)`) as MAX_CRC_DISTANCE
FROM school_details;
"""
pd.read_sql_query(sql, engine)

```

# DISTINCT

```python
sql = """SELECT DISTINCT LocatedInRuralOrUrban AS DISTINCT_LOCALIDAD
FROM school_details;
"""
pd.read_sql_query(sql, engine)

```

# GROUP BY, PROM

```python
sql = """SELECT LocatedInRuralOrUrban, AVG(`DistrictHQDistance (in Km)`) AS PROM_AREA
FROM school_details
GROUP BY LocatedInRuralOrUrban;
"""
pd.read_sql_query(sql, engine)

```

# GROUP BY, MAX

```python
sql = """SELECT PostOfficeName, MAX(`DistrictHQDistance (in Km)`) AS MAX_DISTANCIA
FROM school_details
GROUP BY PostOfficeName;
"""
pd.read_sql_query(sql, engine)

```

# GROUP BY, COUNT, ORDER BY

```python
sql = """SELECT CRCName, COUNT(*) AS CANTIDAD
FROM school_details
GROUP BY CRCName
ORDER BY COUNT(*) DESC;
"""
pd.read_sql_query(sql, engine)

```

# GROUP BY, HAVING

```python
sql = """SELECT CRCName, COUNT(*) AS CANTIDAD
FROM school_details
GROUP BY CRCName
HAVING COUNT(*) > 20;
"""
pd.read_sql_query(sql, engine)

```

```python
sql = """ SELECT SubdivisionName, AVG(`DISTRICTHQDistance (in Km)`) AS Avg_HQ_Distance
FROM school_details
GROUP BY SubdivisionName
HAVING Avg_HQ_Distance > 30;
"""
pd.read_sql_query(sql, engine)

```
