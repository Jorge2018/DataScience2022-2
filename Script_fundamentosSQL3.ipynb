```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
school_df = pd.read_csv(r'https://raw.githubusercontent.com/Jorge2018/RepositorioArchivos/main/School_Details.csv')
school_df.head(2)
```

```python
facility = pd.read_csv(r'https://raw.githubusercontent.com/Jorge2018/RepositorioArchivos/main/School_Facility_Details.csv')
facility.head(2)
```

```python
school_df.info()
```

```python
facility.info()
```

*Creacion de bbdd y conection*

```python
# create sqlite engine for a database called schools
engine = create_engine('sqlite:///schools.db', echo=True)
# create connection to engine
conn = engine.connect()
```

*Agregacion de tb a la bbdd recien creada*

```python
# add the dataframe as a table in sqlite
sqlite_table = "school"
school_df.to_sql(sqlite_table, conn, if_exists='fail')
sqlite_table2 = "school_facility"
facility.to_sql(sqlite_table2, conn, if_exists='fail')


```

```python
# Write your SQL query in this string
sql = """
SELECT AVG(`BACDistance (in Km)`) AS avg_bac_distance FROM SCHOOL ;
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()


```

# R2

```python
# Write the SQL query
sql = """SELECT SchoolLevel,AVG(`BACDistance (in Km)`) AS avg_bac_distance 
FROM SCHOOL
GROUP BY SchoolLevel
ORDER BY  AVG(`BACDistance (in Km)`) DESC ;
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

# R3

```python
# Write the SQL query
sql = """SELECT SchoolLevel,AVG(`BACDistance (in Km)`) AS avg_bac_distance 
FROM SCHOOL
GROUP BY SchoolLevel
HAVING AVG(`BACDistance (in Km)`) > 10
 ;
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```

# R4

```python
# Write the SQL query
sql = """SELECT * FROM SCHOOL AS UNO INNER JOIN SCHOOL_FACILITY AS DOS ON
UNO.EstablishmentCode=DOS.EstablishmentCode
LIMIT 2
 ;
 """
# Read the sql query into pandas to view the results
pd.read_sql_query(sql, engine)
```
