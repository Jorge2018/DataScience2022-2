```python
import pandas as pd
import numpy as np
from pandas import DataFrame
import re
import matplotlib.pyplot as plt
df=pd.read_excel(r'/content/catsvdogs.xlsx')
#df.to_csv('dato'+'.csv',index=False)
```

```python
x =df['Percentage of Dog Owners'] 
num_bins = 7
etiqueta=['Dogs']
fig1 = plt.figure(dpi=120)
ax1 = fig1.add_subplot(1,1,1)
ax1.hist(x, num_bins, density=False)
ax1.set_xlabel("Percentage of animals owners")
ax1.set_ylabel("Count")
ax1.legend(etiqueta)
plt.show()
```

```python
z =df['Percentage of Cat Owners']
num_bins2 = 15
etiqueta=['Cats']
fig1 = plt.figure(dpi=120)
ax1 = fig1.add_subplot(1,1,1)
ax1.hist(z, num_bins2, density=False)
ax1.set_xlabel("Percentage of animals owners")
ax1.set_ylabel("Count")
ax1.legend(etiqueta)
plt.show()
```

```python
x =df['Percentage of Dog Owners'] 
z =df['Percentage of Cat Owners']
num_bins = 7 
num_bins2 = 15
etiqueta=['Dogs','Cats']
fig1 = plt.figure(dpi=120)
ax1 = fig1.add_subplot(1,1,1)
ax1.hist(x, num_bins, density=False)
ax1.hist(z, num_bins2, density=False)
ax1.set_xlabel("Percentage of animals owners")
ax1.set_ylabel("Count")
ax1.legend(etiqueta)
plt.show()
```

```python
data = df['Mean Number of Cats'],df['Mean Number of Dogs per household']
labels = ['Cats', 'Dogs']
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 4))
bplot = axes.boxplot(data,
                         notch=True,
                         vert=True,
                         patch_artist=True,
                         labels=labels)

colors = ['yellow', 'purple']
for bplotA in (bplot,bplot):
    for patch, color in zip(bplotA['boxes'], colors):
        patch.set_facecolor(color)
plt.grid()
plt.show()
print("Esto demuestra que la cantidad de gatos es mayor en comparacion a la cantidad de perros")
```
