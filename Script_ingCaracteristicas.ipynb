```python
import pandas as pd
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vReZBM5OC6GLYbacisp_ToNiu3CLWxqPXw7mWBsdRjnYOFLWNufdQ4qd8u5qTzUF2_sBUAMEi5cgy1U/pub?gid=1040198428&single=true&output=csv')
df.head()
```

*Crea total de integrantes de la familia sumando ambas cols*

```python
df['TotalFamily'] = df['SibSp'] + df['Parch'] 
df = df.drop(['SibSp', 'Parch'], axis=1)
df.head()
```

*redondea valor de edad sin decimales*
*El argumento de Series.round(#) es el número de decimales al que se redondea el número.  Queremos redondearlo a 10, por lo que utilizaremos un número negativo.*

```python
df['Age'] = df['Age'].round(-1)
df.head()
```

*Crea la col GenderAge, cambiando el type de la col Age para luego concatenarla con la col Sex*

```python
df['GenderAge'] = df['Sex'] + df['Age'].astype('string')
df.drop(columns=['Sex','Age'], inplace=True)
df.head()
```

#Características de la cuadratura y multiplicación

*Hagamos una cosa más.  Digamos que queremos normalizar los precios que los pasajeros pagaron.  Decidimos que la manera para hacer esto es multiplicando los precios al cuadrado del Pclass.*

```python
df['NormedFare'] = df['Fare'] * df['Pclass']**2
df.drop(columns='Fare', inplace=True)
df.head()
```

*Separacion Split, str strip elimina espacios vacios (similar al trim sql), para luego obtener datos (FirstName desde la posicion 1 en adelante)*

```python
# create 2 new columns, FirstName and LastName by splitting the Name column
df[['LastName','FirstName']] = df['Name'].str.split(',', expand=True)
# drop the 'Name' column
df.drop('Name', axis=1, inplace=True)

df['FirstName'] = df['FirstName'].str.strip()
df.loc[0, 'FirstName']


df.head()
```

*Concatena cadenas*

```python
df['Name'] = df['FirstName'] + ' ' + df['LastName']
df.drop(columns=['LastName','FirstName'], inplace= True)
df.head()

```
