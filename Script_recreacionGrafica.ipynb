```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
from importlib import reload

df= pd.read_csv(r'https://raw.githubusercontent.com/Jorge2018/DataScience2022-2/main/mortgages.csv')


```

```python
filtro_porc=df['Interest Rate']==0.03
filtro_year=df['Mortgage Name']=='30 Year'
df_year30_porc03=df[filtro_porc & filtro_year]
df_year30_porc05=df[~filtro_porc & filtro_year]
#graficar la suma de esta columna Interest Paid
valdf03=df_year30_porc03['Interest Paid'].cumsum()
valdf05=df_year30_porc05['Interest Paid'].cumsum()
```

```python
x = df_year30_porc03['Month']
y = valdf03
y1= valdf05
x1= df_year30_porc05['Month']
plt=reload(plt)
plt.figure(dpi=100)
plt.plot(x,y,c='b')
plt.plot(x1,y1,c='k' )
plt.xticks(rotation=90)
plt.grid()
etiqueta=('30 years 5% APR','30 years 3% APR')
plt.legend(etiqueta,loc=4)
plt.ylabel("Dollars")
plt.xlabel("Month")
plt.title('Cumulative Interes Paid')
plt.show()
```
